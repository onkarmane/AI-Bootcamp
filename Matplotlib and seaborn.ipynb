# Matplotlib and seaborn


**Matplotlib:**
Matplotlib is a powerful data visualization library in Python. It is essential in machine learning for creating various plots and charts to visualize data distributions, patterns, and relationships. Matplotlib provides a flexible and customizable way to display information, aiding in better understanding and analysis.

**Seaborn:**
Seaborn is built on top of Matplotlib and provides a high-level interface for statistical data visualization. It simplifies the process of creating informative and attractive visualizations, making it especially useful for data exploration in machine learning. Seaborn's additional features, such as built-in themes and color palettes, enhance the aesthetics of plots

```python
import matplotlib.pyplot as plt

# Sample data
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
stock_prices_a = [50, 55, 60, 75, 90, 85, 110, 120, 130, 150, 140]
stock_prices_b = [140, 150, 130, 120, 110, 99, 90, 75, 60, 55, 50]
```

## Axis Labels
Axis labels add labels/names to the X and Y axis.

```python
plt.plot(years,stock_prices_a)
plt.xlabel('Year')
plt.ylabel('Stock Prices');
```

### Plotting Multiple Lines

```python
plt.plot(years, stock_prices_a)
plt.plot(years, stock_prices_b)
plt.xlabel('Year')
plt.ylabel('Stock Prices');
```

### Chart title and Legend

```python
plt.plot(years, stock_prices_a)
plt.plot(years, stock_prices_b)

plt.xlabel('Year')
plt.ylabel('Stock Prices')

plt.title("Stock Prices")
plt.legend(['stock_prices_a', 'stock_prices_b'])
```

### Line Markers

```python
plt.plot(years, stock_prices_a, marker='o')
plt.plot(years, stock_prices_b, marker='x')

plt.xlabel('Year')
plt.ylabel('Stock Prices')

plt.title("Stock Prices")
plt.legend(['stock_prices_a', 'stock_prices_b'])
```

### Styling lines and markers

The `plt.plot` function supports many arguments for styling lines and markers:

- `color` or `c`: set the color of the line ([supported colors](https://matplotlib.org/3.1.0/gallery/color/named_colors.html))
- `linestyle` or `ls`: choose between a solid or dashed line
- `linewidth` or `lw`: set the width of a line
- `markersize` or `ms`: set the size of markers
- `markeredgecolor` or `mec`: set the edge color for markers
- `markeredgewidth` or `mew`: set the edge width for markers
- `markerfacecolor` or `mfc`: set the fill color for markers
- `alpha`: opacity of the plot


```python
plt.plot(years, stock_prices_a, marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(years, stock_prices_b, marker='o', c='r', ls='--', lw=3, ms=10, alpha=.5)

plt.xlabel('Year')
plt.ylabel('Stock Prices')

plt.title("Stock Prices")
plt.legend(['stock_prices_a', 'stock_prices_b'])
```

```python
plt.figure(figsize=(12, 6))
plt.plot(years, stock_prices_a, marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(years, stock_prices_b, marker='o', c='r', ls='--', lw=3, ms=10, alpha=.5)

plt.xlabel('Year')
plt.ylabel('Stock Prices')

plt.title("Stock Prices")
plt.legend(['stock_prices_a', 'stock_prices_b'])
```

## Bar Plot

```python
# Sample data
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
stock_prices_a = [50, 55, 60, 75, 90, 85, 110, 120, 130, 150, 140]
stock_prices_b = [140, 150, 130, 120, 110, 99, 90, 75, 60, 55, 50]
```

```python
plt.bar(years, stock_prices_a);
```

Bars can also be stacked on top of one another. We use the `bottom` argument to `plt.bar` to achieve this.

```python
plt.bar(years, stock_prices_a)
plt.bar(years, stock_prices_b, bottom=stock_prices_a);
```
