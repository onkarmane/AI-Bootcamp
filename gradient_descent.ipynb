```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(20)
Y = (X-5)**2
print(X,Y)

x = 0
lr = 0.1
error = []

plt.plot(X,Y,color='Green')

for i in range(50):
    grad = 2*(x-5)
    x = x - lr*grad
    y = (x-5)**2
    # error.append(y)
    plt.scatter(x,y)
# print(error)
```
